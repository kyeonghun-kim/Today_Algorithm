       . (         .)           (N)   .                .



###

      T .              N, M (0 < N  M < 30) .

            .

```python
N = 5
r = 1
dp = [0] * (N + 1)

for i in range(N+1):
    if i == 0 or i == 1:
        dp[i] = 1
    else:
        dp[i] = i * dp[i-1]

print(dp[N] // (dp[r] * dp[(N-r)]))
print(dp)
```

# BackJoon 9461  
[  ](https://www.acmicpc.net/problem/9461)

### 
---
       .      1.        .      k  ,    k  .

  P(N)     . P(1) P(10)  10  1, 1, 1, 2, 2, 3, 4, 5, 7, 9.

N  , P(N)   .



### 
---
     T .       , N . (1  N  100)

### 
---
   P(N) .

```python
T = 2
N = 6
# P(1) P(10)  10  0 1, 1, 1, 2, 2, 3, 4, 5, 7, 9

for _ in range(T):
    N = int(input())
    dp = [0] * (N + 1)
    for i in range(1, N+1):
        if i == 1 or i == 2:
            dp[i] = 1
        else:
            dp[i] = dp[i-2] + dp[i-3]
    print(dp[N])
```
